A frequency distribution for the outcomes of an experiment. A frequency distribution records the number of times each outcome of an experiment has occurred. For example, a frequency distribution could be used to record the frequency of each word type in a document. Formally, a frequency distribution can be defined as a function mapping from each sample to the number of times that sample occurred as an outcome.

Frequency distributions are generally constructed by running a number of experiments, and incrementing the count for a sample every time it is an outcome of an experiment. For example, the following code will produce a frequency distribution that encodes how often each word occurs in a text:

```python
!pip install nltk
import nltk
nltk.download('all')
```

```python
#http://www.nltk.org/api/nltk.html?highlight=freqdist
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
sent = 'This is an example sentence of another a sentence which example is a sentence'
fdist = FreqDist()
for word in word_tokenize(sent):
  fdist[word.lower()] += 1
```

```python
print(fdist.freq('another'))
'''
Return the frequency of a given sample. 
The frequency of a sample is defined as the count of that sample divided by the total number of sample outcomes
that have been recorded by this FreqDist. The count of a sample is defined as the number of times that sample 
outcome was recorded by this FreqDist. Frequencies are always real numbers in the range [0, 1].
'''
```

```python
fdist.N()
#Return the total number of sample outcomes that have been recorded by this FreqDist.
```

**What is Collocations?**
Collocations are the pairs of words occurring together many times in a document. It is calculated by the number of those pair occurring together to the overall word count of the document.

Consider electromagnetic spectrum with words like ultraviolet rays, infrared rays.

The words *ultraviolet* and *rays* are not used individually and hence can be treated as Collocation. Another example is the *CT Scan*. We don't say CT and Scan separately, and hence they are also treated as collocation.

We can say that finding collocations requires calculating the frequencies of words and their appearance in the context of other words. These specific collections of words require filtering to retain useful content terms. Each gram of words may then be scored according to some association measure, to determine the relative likelihood of each Ingram being a collocation.

Thus *lions roar* is a collocation, whereas *lions shout* is not. A characteristic of collocations is that they are resistant to substitution with words that have similar senses; for example, *blow up in tears* sounds definitely odd compared to  *burst into tears*

To create a **Words Phrase Frequency Counter** we can use N-Grams, where a phrase can be a bigram, trigram and so on. 

```python
from nltk.util import ngrams
def wordPhraseFrequencyCounter(sent_Text,no_of_words_in_phrase):
  phrases = ngrams(nltk.word_tokenize(sent_Text), no_of_words_in_phrase)
  f__dist_of_phrase = FreqDist()
  for prase in phrases:
    #("as","good","as") ---> "ad adad adad"
    f__dist_of_phrase[" ".join(prase).lower()] += 1
  print(f__dist_of_phrase.N())
  print(f__dist_of_phrase.tabulate())

#compare results with http://www.writewords.org.uk/phrase_count.asp
```

How is Collocations different than regular BiGrams or TriGrams?
The set of two words that co-occur as BiGrams, and the set of three words that co-occur as TriGrams, may not give us meaningful phrases. For example, the sentence ‘He applied machine learning’ contains bigrams: ‘He applied’, ‘applied machine’, ‘machine learning’. ‘He applied’ and ‘applied machine’ do not mean anything, while ‘machine learning’ is a meaningful bigram. Just considering co-occurring words may not be a good idea, since phrases such as ‘of the’ may co-occur frequently, but are actually not meaningful. 

```python
import nltk
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
trigram_measures = nltk.collocations.TrigramAssocMeasures()
finderbi = BigramCollocationFinder.from_words(nltk.word_tokenize("The seven islands that constitute Mumbai were lions roar originally home to communities of Marathi language speaking Koli people are in the sample example of lions roar  and are in the sample example of a rush of anger and again in place of a rush of anger"))
print("bigram",finderbi.nbest(bigram_measures.pmi, 10))
finderti = TrigramCollocationFinder.from_words(nltk.word_tokenize("The seven islands that constitute Mumbai were lions roar originally home to communities of Marathi language speaking Koli people are in the sample example of lions roar  and are in the sample example of a rush of anger and again in place of a rush of anger"))
print("trigram",finderti.nbest(trigram_measures.pmi, 10))

for i in finderti.score_ngrams(trigram_measures.likelihood_ratio):
    print (i)
```

**Word cloud** is a technique for visualising frequent words in a text where the size of the words represents their frequency.

Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance. Significant textual data points can be highlighted using a word cloud. Word clouds are widely used for analyzing data from social network websites.

Word Clouds are a popular way of displaying how important words are in a collection of texts. Basically, the more frequent the word is, the greater space it occupies in the image. One of the uses of Word Clouds is to help us get an intuition about what the collection of texts is about. Here are some classic examples of when Word Clouds can be useful:

*   Take a quick peek at the word distribution of a collection of texts
*   Clean the texts and want to see what are some frequent stopwords you want to filter out
*   See the differences between frequent words between two or more collections of texts

Let’s suppose you want to build a text classification system. If you’d want to see what are the different frequent words in the different categories, you’d build a Word Cloud for each category and see what are the most popular words inside each category.

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize 
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
l = list(stopwords.words('english'))
print(l)
wikipedia.set_lang("en")
result= wikipedia.summary("Mumbai", sentences=10)
word_tokens = word_tokenize(result) 
filtered_sentence = []  
  
for w in word_tokens:  
    if w not in l:  
        filtered_sentence.append(w)  
print(' '.join(filtered_sentence))
lemma_sentence = []
for f in filtered_sentence:
    a = lemmatizer.lemmatize(f)
    lemma_sentence.append(a)
r = ' '.join(lemma_sentence)
print(r)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
def plot_cloud(wordcloud):
    plt.figure(figsize=(8, 8))
    plt.imshow(wordcloud) 
    plt.axis("off");
wordcloud = WordCloud(width = 500, height = 500, background_color='black',colormap='Set2',random_state=5).generate(r)
plot_cloud(wordcloud)
```
